```python
import numpy as np
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('data.csv', parse_dates=['date'])
```

```python
data.head()
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(data.drop('date', axis=1).corr())
```

It appears that the sales are not sufficiently correlated with the price of fuel. However there might be some value in predicting the (weighted) price of cheddar based on the prices of fuel.

```python
plt.figure(figsize=(12,8))
ax = sns.lineplot(data=data, x='date', y='weighted_prices')
ax.set_title('weighted_prices Time Series Plot')
ax.set_xlabel('Date')
ax.set_ylabel('Weighted Prices ($)')

plt.show()
```

```python
plt.figure(figsize=(12,8))
ax = sns.lineplot(data=data, x='date', y='sales')
ax.set_title('Sales Time Series Plot')
ax.set_xlabel('Date')
ax.set_ylabel('N.o. Sales')

plt.show()
```

```python
#sales based on day of week
data['date'].dt.day_name().value_counts()
```

```python
plt.figure(figsize=(12,8))
ax = sns.boxplot(x=data['date'].dt.day_name(), y='sales', data=data)
ax.set_title('Sales Per Day Of Week')
ax.set_xlabel('Date')
ax.set_ylabel('N.o. Sales')

plt.show()
```

```python
plt.figure(figsize=(12,8))
ax = sns.boxplot(x=data['date'].dt.day_name(), y='weighted_prices', data=data)
ax.set_title('Weighted Prices Per Day Of Week')
ax.set_xlabel('Date')
ax.set_ylabel('Weighted Prices ($)')

plt.show()
```
