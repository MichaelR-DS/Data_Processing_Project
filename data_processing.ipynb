```python
import numpy as np
import pandas as pd
from datetime import datetime as dt
```

[From the description:](https://www.kaggle.com/mruanova/us-gasoline-and-diesel-retail-prices-19952021) The data contains the following information:

* A1: Weekly U.S. All Grades All Formulations Retail Gasoline Prices (Dollars per Gallon)
* A2: Weekly U.S. All Grades Conventional Retail Gasoline Prices (Dollars per Gallon)
* A3: Weekly U.S. All Grades Reformulated Retail Gasoline Prices (Dollars per Gallon)
* R1: Weekly U.S. Regular All Formulations Retail Gasoline Prices (Dollars per Gallon)
* R2: Weekly U.S. Regular Conventional Retail Gasoline Prices (Dollars per Gallon)
* R3: Weekly U.S. Regular Reformulated Retail Gasoline Prices (Dollars per Gallon)
* M1: Weekly U.S. Midgrade All Formulations Retail Gasoline Prices (Dollars per Gallon)
* M2: Weekly U.S. Midgrade Conventional Retail Gasoline Prices (Dollars per Gallon)
* M3: Weekly U.S. Midgrade Reformulated Retail Gasoline Prices (Dollars per Gallon)
* P1: Weekly U.S. Premium All Formulations Retail Gasoline Prices (Dollars per Gallon)
* P2: Weekly U.S. Premium Conventional Retail Gasoline Prices (Dollars per Gallon)
* P3: Weekly U.S. Premium Reformulated Retail Gasoline Prices (Dollars per Gallon)
* D1: Weekly U.S. No 2 Diesel Retail Prices (Dollars per Gallon)

```python
gas_data = pd.read_csv('/Users/michael/Desktop/data_cleaning_project/data/U.S. Gasoline and Diesel Retail Prices 1995-2021.csv', parse_dates= ['Date'])

cheddar_data = pd.read_csv('/Users/michael/Desktop/data_cleaning_project/data/Weekly Dairy Product Prices/Datamart-Export_DY_WK100-40 Pound Block Cheddar Cheese Prices and Sales_20170829_122601.csv')
```

cheddar data

```python
cheddar_data.head()
```

```python
cheddar_data['Report Date'] = pd.to_datetime(cheddar_data['Report Date'], format='%m/%d/%Y').dt.strftime('%d/%m/%Y')
```

```python
cheddar_data['Date'].to_string()
cheddar_data['day'] = cheddar_data['Date'].apply( lambda x: x.split('/')[1] )
cheddar_data['month'] = cheddar_data['Date'].apply( lambda x: x.split('/')[0] )
cheddar_data['year'] = pd.DatetimeIndex(cheddar_data['Report Date']).year
```

```python
cheddar_data.drop(['Report Date', 'Date', 'Week Ending Date'], axis=1, inplace=True)
```

```python
cheddar_data.head()
```

```python
cheddar_data['date'] = pd.to_datetime(cheddar_data[['day', 'month', 'year']], format='%d/%m/%Y', errors='coerce')
```

```python
cheddar_data
```

```python
cheddar_data = cheddar_data.drop(['day', 'month', 'year'], axis=1)
```

```python
cheddar_start_date = min(cheddar_data['date'])
cheddar_latest_date =  max(cheddar_data['date'])
```

```python
gas_data['Date'] = pd.to_datetime(gas_data['Date'], format='%d/%m/%Y')
```

```python
#get dates within same range as cheddar_data
gas_data_revised = gas_data[(gas_data['Date'] >= cheddar_start_date) & (gas_data['Date'] <= cheddar_latest_date)]
```

```python
#make sure two date columns have same name
gas_data_revised.rename({'Date':'date'}, axis=1, inplace=True)
```

```python
gas_data_revised
```

```python
cheddar_data.sort_values(['date'], axis=0, ascending=True, inplace=True)
gas_data_revised.sort_values(['date'], axis=0, ascending=True, inplace=True)
```

```python
data = pd.merge_asof(cheddar_data, gas_data_revised, on='date', direction='nearest')
```

```python
data.isnull().any().any()
```

```python
data.rename({'Weighted Prices':'weighted_prices', 'Sales':'sales'}, axis='columns', inplace=True)
```

```python
#making 'sales' column numeric
data['sales'].replace(',','', regex=True, inplace=True)
```

```python
data
```

```python
data['date'] = pd.to_datetime(data['date'], format='%d/%m/%Y')
```

```python
 data['date'].dt.strftime('%d/%m/%Y')
```

```python
data.to_csv('data.csv', index=False, date_format='%d/%m/%Y')
```
